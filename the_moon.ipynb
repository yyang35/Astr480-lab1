## The Moon effect on 1st May
What will the moon phase be during your observations? Will the moon interfere with your observations? Explain your reasoning:

**Finding the Moon phase and illumination**

```python
from astropy.coordinates import get_sun, get_body, get_moon
from astroplan import moon_illumination
from astroplan import moon_phase_angle
from astropy.time import Time
from astroplan import FixedTarget, Observer
import astropy.units as u
```

```python
mcdonald_observatory = Observer(longitude = -104.022 * u.deg,
                    latitude = 30.6714 * u.deg,
                    elevation = 2077 * u.m,
                    timezone = 'US/Central',
                    name = "McDonald Observatory"
                    )
```

```python
reference_time = Time("2023-05-01")

midnight = mcdonald_observatory.midnight(reference_time, which='next')
astro_set = mcdonald_observatory.twilight_evening_astronomical(reference_time, which='nearest')
astro_rise = mcdonald_observatory.twilight_morning_astronomical(reference_time, which='next')
```

```python
illumination = moon_illumination(midnight)
phase = moon_phase_angle(midnight)
```

```python
print(f"In may 1st, the moon has a phase of {phase:.2f}, and with a illumination of {illumination:.2f}")
```

**Plot the moon location at the night of May 1st**\
Use get_moon() method to get the coordination of moon at may 1st, then use the plot_sky method to show up the loaction of moon at the night of May 1st.

```python
moon_coord = get_moon(midnight)
moon = FixedTarget(name = 'moon', coord = moon_coord)
```

```python
from astroplan import time_grid_from_range
from astroplan.plots import plot_sky
import matplotlib.pyplot as plt

observing_range = [astro_set, astro_rise]
time_grid = time_grid_from_range(observing_range)
moon_style = {'color': 'w','marker': 'o','edgecolors' : 'k','linewidth' : 2}
ax = plot_sky(moon, mcdonald_observatory, time_grid ,style_kwargs = moon_style)
```

**Conclusion**\
The moon effect at May 1st is realtive large, since it is in a small radian of phase, so it near full moon, and having a 0.8 illumination which is large, and as the plot shows the moon is at high altitude for the whole night, which probably could effect a large amount of sky.
